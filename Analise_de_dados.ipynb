```python
#Visualiza as duas colunas de data e moeda
df = df[['date', 'currency']]
df.head(10)
```

```python
#Mostra os dados ordenados
df.sort_values('currency', ascending=False).head(10)
```

```python
#Esse código cria um gráfico de linha que mostra a quantidade de registros em cada ano.
df['date'] = pd.to_datetime(df['date'])

year_counts = df['date'].dt.year.value_counts().sort_index()

plt.figure(figsize=(10,6))
plt.plot(
    year_counts.index,
    year_counts.values,
    marker='o'
)

plt.title('Registros por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade')

plt.show()
```

```python
#Esse código cria um gráfico de pizza mostrando a distribuição das moedas mais frequentes na coluna
import matplotlib.pyplot as plt

moedas = df['currency'].value_counts().head(10)

plt.figure(figsize=(8,8))
plt.pie(moedas.values,
        labels=moedas.index,
        autopct='%1.1f%%')

plt.title('Distribuição das Moedas')
plt.show()
```
